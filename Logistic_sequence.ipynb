```python
import numpy as np

def p_curr_logistic(p_prev,alpha):
    p_curr=alpha*p_prev*(1-p_prev)
    return p_curr

def compute_logistic(p0,alpha,n_max):
  n_vec=np.linspace(2,n_max,n_max-1)
  p_n_vec=[]
  p_prev=p0
  p_n_vec.append(p0)
  for n in n_vec:
      p_curr=p_curr_logistic(p_prev,alpha)
      p_n_vec.append(p_curr)
      p_prev=p_curr
  return p_n_vec

p0=1/3
alpha=2
n_max=100

p_n_vec=compute_logistic(p0,alpha,n_max)
print(p_n_vec)
```

```python
import matplotlib.pyplot as plt

def plot_logistic(p0,alpha,n_max):
    n_vec=np.linspace(1,n_max,n_max)
    p_n_vec=compute_logistic(p0,alpha,n_max)
    fig, ax = plt.subplots(figsize=(20, 6))
    ax.plot(n_vec, p_n_vec,"o")
    ax.set(xlabel='x',ylabel='y')
    ax.axhline(y=0,color="k")
    ax.axvline(x=0,color="k")
    ax.grid(True)
    plt.show()
```

a) $p_0=1/2$ and $\alpha$ between 1 and 3

```python
p0=1/2
alpha=1.5
n_max=50

print(compute_logistic(p0,alpha,n_max))
plot_logistic(p0,alpha,n_max)
```

```python
p0=1/2
alpha=2.5
n_max=50

print(compute_logistic(p0,alpha,n_max))
plot_logistic(p0,alpha,n_max)
```
